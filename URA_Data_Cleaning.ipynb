# Cleaning Data from Urban Redevelopment Authority API

## Concepts 
- show how to extract values from lists and dictionaries in JSON file 
- show how to automate munging with a function 
- write data frame to CSV
- use command line magic commands 
- ask questions on Stack Overflow

```python
import pandas as pd 
import json
```

```python
with open('PropertyTransactions1.json') as f: 
    json_data = f.read() 
    prop = json.JSONDecoder().decode(json_data)
```

```python
type(prop)
```

```python
prop.keys()
```

```python
result = prop['Result']
type(result)
```

```python
prop['Result'][0]
```

```python
ls_one = []
for result in prop['Result']:
    if 'x' in result.keys():
        ls_one.append(result)
    else:
        continue
```

```python
housing_1 = pd.DataFrame()
for result in ls_one:
    transact = result['transaction']
    transact_df = pd.DataFrame(transact)
    transact_df['marketSegment'] = result['marketSegment']
    transact_df['street'] = result['street']
    transact_df['project'] = result['project']
    transact_df['x'] = result['x']
    transact_df['y'] = result['y']
    housing_1 = housing_1.append(transact_df)
```

```python
housing_1.head(20)
```

```python
housing_1.to_csv('housing_1.csv', sep=',')
```
