# Visualisation Principles with the Boston Housing Dataset 

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
# sklearn boston dataset is a dictionary so explore keys of dictionary 
boston.keys()
```

```python
# convert data values into data frame to make it easier to work with 
bos = pd.DataFrame(boston.data)
bos.head()
```

```python
# replace column names with feature names 
bos.columns = boston.feature_names
bos.head()
```

# Viewing Distributions with Histograms, Density Plots and Box-and-Whisker Plots 
## Univariate Histograms 

```python
bos.hist()
```

# Univariate Density Plots 

```python
bos.plot(kind='density', subplots=True, layout=(3,5), sharex=False)
```

```python
bos.plot(kind='box', subplots=True, layout=(5,3), sharex=False, sharey=False)
```

## Scatterplot Matrix

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(bos)
```

```python
# calculate correlations between all variables 
corr = bos.corr()
```

```python
# set up matplot lib figure 
f, ax = plt.subplots(figsize=(12,9))

# draw heatmap with seaborn 
sns.heatmap(corr, vmax=.8, square=True)
```
